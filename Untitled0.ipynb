```python
from sklearn.datasets import fetch_20newsgroups
```

```python
text_data = fetch_20newsgroups()
```

```python
import numpy as np
```

```python
raw_text = text_data.data[:2]
raw_text
```

Conversion to lowercase

```python
s1 = []
def tolower(a, b):
  for words in a:
    b.append(str.lower(words))
```

```python
tolower(raw_text, s1)
```

```python
s1
```

Tokenization

```python
s2 = []
from nltk.tokenize import sent_tokenize, word_tokenize

import nltk
nltk.download('punkt')
```

```python
a = []
def sentence_tok(a, b):
  for sentence in a:
    sentence = sent_tokenize(sentence)
    b.append(sentence)
```

```python
sentence_tok(s1, a)
a
```

```python
def word_tok(a, b):
  for i in a:
    i = word_tokenize(i)
    b.append(i)
```

```python
word_tok(s1, s2)
s2
```

Removing Punctuation

```python
import re
s3 = []
for words in s2:
  clean = []
  for w in words:
    res = re.sub(r'[^\w\s]',"",w)
    if res != "":
      clean.append(res)
    s3.append(clean)
```

```python
s3
```

Removing Stop Words

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
s4 = []
for i in s3:
  m = []
  for j in i:
    if not j in stopwords.words('english'):
      m.append(j)
  s4.append(m)
```

```python
s4
```

Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
port = PorterStemmer()
s5 = []
```

```python
for i in s4:
  w = []
  for j in i:
    w.append(port.stem(j))
  s5.append(w)
```

```python
s5
```

Lemitization

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
wnet = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
l = []
for i in s4:
  w = []
  for j in i:
    w.append(wnet.lemmatize(j))
  l.append(w)
```
